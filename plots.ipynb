```python
import matplotlib.pyplot as plt

cores = [2, 4, 6, 8, 10]
```

```python
## Plot SS FPGrowth
total_time_SS_FP = [80.6, 68.5, 44.1, 42.1, 40.4]
data_read_SS_FP = [28.3, 23.5, 16.5 , 15.5, 14.7]

plt.title('Strong Scaling FP-Growth')
plt.plot(cores, total_time_SS_FP, '-', label='Total')
plt.plot(cores, data_read_SS_FP, '--', label='HDFS data read')

plt.xlabel('Cores')
plt.ylabel('Runtime (seconds)')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
## Plot WS FPGrowth
total_time_WS_FP = [80.6, 65.0, 66.0, 66.3, 77.6]
data_read_WS_FP = [28.3, 19.1, 17.3, 15.1, 14.5]

plt.title('Weak Scaling FP-Growth')
plt.plot(cores, total_time_WS_FP, '-', label='Total')
plt.plot(cores, data_read_WS_FP, '--', label='HDFS data read')

plt.xlabel('Cores')
plt.ylabel('Runtime (seconds)')
plt.legend(loc='best')
plt.grid()
plt.show()
```
